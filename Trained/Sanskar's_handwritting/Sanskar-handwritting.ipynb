```python
import cv2
import easyocr
import matplotlib.pyplot as plt
reader = easyocr.Reader(['en'])
filename = 'Dataset/11.png'
img = cv2.imread(filename)

results = reader.readtext(img)

recognized_text = ""

for (bbox, text, prob) in results:
    # Draw bounding box (optional)
    (top_left, top_right, bottom_right, bottom_left) = bbox
    top_left = tuple(map(int, top_left))
    bottom_right = tuple(map(int, bottom_right))
    cv2.rectangle(img, top_left, bottom_right, (0, 255, 0), 2)

    recognized_text += text + " "
    
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.title("Detected Characters")
plt.show()

print("Recognized Text:", recognized_text)
```

```python
import cv2
import easyocr
import matplotlib.pyplot as plt

# Initialize the EasyOCR reader
reader = easyocr.Reader(['en'])

# List of image file paths
filenames = ['Dataset/Bishal/01.jpg', 'Dataset/Bishal/02.jpg', 'Dataset/Bishal/03.jpg']

# Loop through each file
for filename in filenames:
    try:
        # Read the image file
        img = cv2.imread(filename)
        
        # Check if the image is loaded successfully
        if img is None:
            print(f"Error: Unable to load image at {filename}")
            continue

        # Run OCR on the image using EasyOCR
        results = reader.readtext(img)

        recognized_text = ""

        # Loop through OCR results
        for (bbox, text, prob) in results:
            # Draw bounding box (optional)
            (top_left, top_right, bottom_right, bottom_left) = bbox
            top_left = tuple(map(int, top_left))
            bottom_right = tuple(map(int, bottom_right))
            cv2.rectangle(img, top_left, bottom_right, (0, 255, 0), 2)

            # Append recognized text
            recognized_text += text + " "

        # Convert BGR to RGB for display
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Display the image with bounding boxes
        plt.imshow(img_rgb)
        plt.axis('off')
        plt.title(f"Detected Characters in {filename}")
        plt.show()

        # Print the recognized text
        print(f"Recognized Text from {filename}: {recognized_text}")

    except Exception as e:
        print(f"Error processing {filename}: {e}")
```
